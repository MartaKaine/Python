```python
def menu():
    options = int(input("Please select option:\n1 - Display balance\n2 - Withdraw\n3 - Deposit\n4 - Exit\n"))
    return options

def show_balance(balance):
    print(f"Your current balance: {balance}")

def sub_menu():
    options = int(input("Please choose withdrawal amount:\n1 - £10\n2 - £20\n3 - £50\n4 - other\n"))
    return options

def withdrawal(balance):
    while True:
        choice = sub_menu()
        amount = 0
        if choice == 1:
            amount = 10
        elif choice == 2:
            amount = 20
        elif choice == 3:
            amount = 50
        elif choice == 4:
            while True :
                amount = int(input("Please enter withdrawal amount: "))
                if amount > balance:
                    print("Insufficient funds.")
                    return balance
                if amount % 10 != 0 :
                    print("Please enter a multiple of 10.")
                    continue
                break
        else:
            print("Invalid input.")
            continue
        if amount > balance:
            print("Insufficient funds.")
            return balance

        print("Transaction successful.")
        balance -= amount
        return balance

def deposit(balance):
    amount = int(input("Please enter the amount you wish to deposit and insert the notes: "))
    balance += amount
    print("Transaction complete.")
    return balance

def exit_atm():
    print("Goodbye!")

def pin_validation():
    pin_no = int(input("Please enter PIN: "))
    return pin_no

def ATM():
    balance = 500
    pin = 1234
    while True:
        pin_entered = pin_validation()
        if pin_entered == pin:
            print("PIN accepted.")
            break
        else:
            print("Incorrect PIN number.\nPlease try again.")
            continue
    while True:
        choice = menu()
        if choice == 1 :
            show_balance(balance)
            continue
        elif choice == 2:
            balance = withdrawal(balance)
            continue
        elif choice == 3:
            balance = deposit(balance)
            continue
        elif choice == 4:
            exit_atm()
            break
        else:
          print("Invalid input.")
          continue

ATM()
```
